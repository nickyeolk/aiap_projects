## Initial Model

````python
# handle nan values
# print(titanic_raw.isnull().sum())
def preprocessing(df):
    '''Handles nans and drops name, passengerID'''
#     df.dropna(axis=0,subset=['Embarked'],inplace=True) # remove if embarkation is unknown
    df = df.fillna({'Age':np.nanmean(df['Age']) , 'Fare':np.nanmean(df['Fare']), 'Embarked':'S'})
    df = df.drop(['PassengerId','Ticket','Cabin'],axis=1)
    return df
df_train = preprocessing(titanic_raw)
test_id = titest['PassengerId']
df_test = preprocessing(titest)

print(df_train.isnull().sum(),titanic_raw.isnull().sum())
````

````python
# convert strings to numeric categories

def get_dict(df,variable):
    dictor={}
    df[variable] = df[variable].astype('category')
    for ii,item in enumerate(df[variable].cat.categories):
        dictor[item]=ii
    return dictor
    
    
sexdict = get_dict(df_train,'Sex')
embarkeddict = get_dict(df_train,'Embarked')
varlist = ['Sex','Embarked']
dictlist = [sexdict,embarkeddict]

def str_2_cat(df,var,dictionary):
    def func(var,dictionary):
        return dictionary[var]
    df[var] = df.apply(lambda x: func(x[var],dictionary),axis=1)
    return df

# def str_2_cat(df):
# #     df['Name'] = df['Name'].astype('category').cat.codes
#     df['Sex'] = df['Sex'].astype('category').cat.codes
#     df['Ticket'] = df['Ticket'].astype('category').cat.codes
#     df['Cabin'] = df['Cabin'].astype('category').cat.codes
#     df['Embarked'] = df['Embarked'].astype('category').cat.codes
#     return df
for ii,variable in enumerate(varlist):
    df_train = str_2_cat(df_train,variable,dictlist[ii])
    df_test = str_2_cat(df_test,variable,dictlist[ii])
# note. in the future just use .map(dictionary)
````

````python
from sklearn.utils import shuffle
def split_data(df_data,train_ratio,test_ratio,targetvar):
    '''Splits data into train, test sets'''
    df_data=shuffle(df_data)
    df_train=df_data.iloc[:int(train_ratio*len(df_data)),:]
    df_test=df_data.iloc[int(train_ratio*len(df_data)):int((train_ratio+test_ratio)*len(df_data)),:]
    X_train=df_train.drop([targetvar],axis=1).values
    X_test=df_test.drop([targetvar],axis=1).values
    y_train=df_train[targetvar].values
    y_test=df_test[targetvar].values
    
    if((train_ratio+test_ratio)<1):
        df_validate = df_data.iloc[int((train_ratio+test_ratio)*len(df_data)):,:]
        X_validate=df_validate.drop(['Last Price'],axis=1).values
        y_validate=df_validate['Last Price'].values
        return X_train, y_train, X_test, y_test , X_validate, y_validate
    return X_train, y_train, X_test, y_test
````

````python
X_train,y_train,X_test,y_test = split_data(df_train,0.7,0.3,'Survived')
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
rf.score(X_test,y_test)
````

````python
def make_kaggle_submission(train_df,test_df,rf):
    X_train,y_train,X_test,y_test = split_data(train_df,1.0,0.0,'Survived')
    rf.fit(X_train,y_train)
    print(rf.get_params())
    feature_names = test_df.columns
    feature_importance = rf.feature_importances_
    indices = np.argsort(feature_importance)[::-1]
    print(list(zip(feature_names[indices],feature_importance[indices])))
    print(np.shape(X_train),np.shape(y_train))
    prediction = pd.DataFrame({'PassengerId':test_id ,'Survived':rf.predict(test_df)})
    print(len(prediction))
#     print(rf.oob_score_)
    prediction.to_csv('./submission.csv',index=False)
make_kaggle_submission(df_train,df_test,rf)
````

````python
titanic_raw = pd.read_csv('./data/train.csv')
titest = pd.read_csv('./data/test.csv')
def preprocessing(df):
    '''Handles nans and drops name, passengerID'''
#     df.dropna(axis=0,subset=['Embarked'],inplace=True) # remove if embarkation is unknown
    df = df.fillna({'Age':np.nanmean(df['Age']) ,'Fare':np.nanmean(df['Fare']) , 'Embarked':'S', \
                  'Cabin':'None'})
    df = df.drop(['PassengerId','Ticket'],axis=1)
    return df
df_train = preprocessing(titanic_raw)
test_id = titest['PassengerId']
df_test = preprocessing(titest)
sexdict = get_dict(df_train,'Sex')
embarkeddict = get_dict(df_train,'Embarked')
varlist = ['Sex','Embarked']
dictlist = [sexdict,embarkeddict]
for ii,variable in enumerate(varlist):
    df_train = str_2_cat(df_train,variable,dictlist[ii])
    df_test = str_2_cat(df_test,variable,dictlist[ii])


'''New stuff starts here'''
# #Fill fare
# classfares = titanic_raw.groupby('Pclass')['Fare'].mean().values
# def func(df,classfares):
#     if np.isnan(df['Fare']):
#         df['Fare'] = classfares[df['Pclass']-1]
#     else: return df['Fare']
# df_train['Fare'] = df_train.apply(lambda x: func(x,classfares),axis=1)
# df_test['Fare'] = df_test.apply(lambda x: func(x,classfares),axis=1)
        

df_train['Total_fam'] = df_train['SibSp']+df_train['Parch']
df_test['Total_fam'] = df_test['SibSp']+df_test['Parch']
def func(totalfam):
    if totalfam>0:
        return 0
    else: return 1
df_train['Isalone'] = df_train.apply(lambda x: func(x['Total_fam']),axis=1)
df_test['Isalone'] = df_test.apply(lambda x: func(x['Total_fam']),axis=1)
def func(totalfam):
    if((totalfam>0)&(totalfam<4)):
        return 1
    else: return 0
# df_train['Smallfam'] = df_train.apply(lambda x: func(x['Total_fam']),axis=1)
# df_test['Smallfam'] = df_test.apply(lambda x: func(x['Total_fam']),axis=1)
df_train.drop('Total_fam',inplace=True,axis=1)
df_test.drop('Total_fam',inplace=True,axis=1)

#extra stuff
df_train['Name_len'] = titanic_raw['Name'].apply(len)
df_test['Name_len'] = titest['Name'].apply(len)
df_train['has_cabin'] = titanic_raw['Cabin'].apply(lambda x: 0 if type(x)==float else 1 )
df_test['has_cabin'] = titest['Cabin'].apply(lambda x: 0 if type(x)==float else 1 )

# update age
age_avg = titanic_raw['Age'].mean()
age_std = titanic_raw['Age'].std()

age_null_count_train = titanic_raw['Age'].isnull().sum()
age_null_count_test = titest['Age'].isnull().sum()
age_null_random_values_train = np.random.randint(age_avg - age_std, age_avg + age_std, size=age_null_count_train)
age_null_random_values_test = np.random.randint(age_avg - age_std, age_avg + age_std, size=age_null_count_test)
df_train['Age'][np.isnan(titanic_raw['Age'])] = age_null_random_values_train
df_train['Age'] = df_train['Age'].astype(int)
df_test['Age'][np.isnan(titest['Age'])] = age_null_random_values_test
df_test['Age'] = df_test['Age'].astype(int)

# group ages
def categorize_ages(df):
    df['Age_group'] = df['Age']
    df['Age_group'].loc[df['Age']<15] = 0
    df['Age_group'].loc[(df['Age']>=15) & (df['Age']<30)] = 1
    df['Age_group'].loc[(df['Age']>=30) & (df['Age']<65)] = 2
    df['Age_group'].loc[df['Age']>=65] = 3
    return df['Age_group']
# df_train['Age_group'] = categorize_ages(df_train)
# df_test['Age_group'] = categorize_ages(df_test)

# titles
def make_namedict(df):
    df['Title'] = df.Name.str.extract(' ([A-Za-z]+)\.', expand=False)
    df['Title'] = df['Title'].replace(['Lady', 'Countess','Capt', 'Col',\
    'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
    df['Title'] = df['Title'].replace('Mlle', 'Miss')
    df['Title'] = df['Title'].replace('Ms', 'Miss')
    df['Title'] = df['Title'].replace('Mme', 'Mrs')
    namedict = get_dict(df,'Title')
    return namedict
namedict = make_namedict(df_train)
def insert_title(df,namedict):
    df['Title'] = df.Name.str.extract(' ([A-Za-z]+)\.', expand=False)
    df['Title'] = df['Title'].replace(['Lady', 'Countess','Capt', 'Col',\
    'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
    df['Title'] = df['Title'].replace('Mlle', 'Miss')
    df['Title'] = df['Title'].replace('Ms', 'Miss')
    df['Title'] = df['Title'].replace('Mme', 'Mrs')
    df = str_2_cat(df,'Title',namedict)
    return df['Title']

df_train['Title'] = insert_title(df_train,namedict)
df_test['Title'] = insert_title(df_test,namedict)

#cabin code
def make_cabindict(df):
    df['Cabin_code'] = df['Cabin'].str[0]
    cabindict = get_dict(df_train,'Cabin_code')
    return cabindict
cabindict = make_cabindict(df_train)
def input_cabin_code(df,cabindict):
    df['Cabin_code'] = df['Cabin'].str[0]
    df = str_2_cat(df,'Cabin_code',cabindict)
    return df['Cabin_code']
df_train['Cabin_code'] = input_cabin_code(df_train,cabindict)
df_test['Cabin_code'] = input_cabin_code(df_test,cabindict)


df_train.drop(['Name','Cabin'],inplace=True, axis=1)
df_test.drop(['Name','Cabin'],inplace=True, axis=1)
````

### Gini Criterion

If you remember from `get_params`, there exists a parameter `criterion: 'gini'`. This means that the tree is using Gini as a criterion to decide how to separate the data.

Hence, we will first learn how to use the Gini impurity score. The Gini impurity score of a node n is given as:  

<center>$i(n) = 1 - p^2_0 - p^2_1$,  </center>  

Where $p_1$ refers to the proportion of 1's in that node, and $p_0$ refers to the proportion of 0's.

````python
def infogain(a,b):
    def gaincalc(array):
        positives = sum(array)
        total = len(array)
        fra=(positives/total)+1e-3
        print('fra',fra)
        return -fra*np.log2(fra)-(1.-fra)*np.log2(1.-fra)
    w1 = len(a)/(len(a)+len(b))
    w2 = 1-w1
    return gaincalc(a)*w1 + gaincalc(b)*w2
````

## Random Forests

Now that we have a decision tree, we can build a random forest, comprising of decision trees of randomised bootstraps of our dataset. At the simplest level, a random forest can be simply a list of decision trees that take a vote on the outcome of the prediction. This list can be an attribute of the random forest.

The basic modification of random forests is the use of bootstrapping. Bootstrapping is done in a few lines of code through `np.random.choice`.

Hence, to begin, build a simple random forest, that will initialise 5 trees through bootstrapping (sampling 100% with replacement), and predict the answer through a voting mechanism out of all the 5 trees. For computational efficiency, we recommend using `np.stack` and `np.array.mean`.

````python
# -*- coding: utf-8 -*-
"""
Created on Sat Nov 17 15:29:56 2018

@author: likkhian
"""

import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.cross_validation import train_test_split
from sklearn.datasets import make_hastie_10_2
import matplotlib.pyplot as plt

""" HELPER FUNCTION: GET ERROR RATE ========================================="""
def get_error_rate(pred, Y):
    return sum(pred != Y) / float(len(Y))

""" HELPER FUNCTION: PRINT ERROR RATE ======================================="""
def print_error_rate(err):
    print('Error rate: Training: %.4f - Test: %.4f' % err)

""" HELPER FUNCTION: GENERIC CLASSIFIER ====================================="""
def generic_clf(Y_train, X_train, Y_test, X_test, clf):
    clf.fit(X_train,Y_train)
    pred_train = clf.predict(X_train)
    pred_test = clf.predict(X_test)
    return get_error_rate(pred_train, Y_train), \
           get_error_rate(pred_test, Y_test)
    
""" ADABOOST IMPLEMENTATION ================================================="""
def adaboost_clf(Y_train, X_train, Y_test, X_test, M, clf):
    n_train, n_test = len(X_train), len(X_test)
    # Initialize weights
    w = np.ones(n_train) / n_train
    pred_train, pred_test = [np.zeros(n_train), np.zeros(n_test)]
    
    for i in range(M):
        # Fit a classifier with the specific weights
        clf.fit(X_train, Y_train, sample_weight = w)
        pred_train_i = clf.predict(X_train)
        pred_test_i = clf.predict(X_test)
        # Indicator function
        miss = [int(x) for x in (pred_train_i != Y_train)]
        # Equivalent with 1/-1 to update weights
        miss2 = [x if x==1 else -1 for x in miss]
        # Error
        err_m = np.dot(w,miss) / sum(w)
        # Alpha
        alpha_m = 0.5 * np.log( (1 - err_m) / float(err_m))
        # New weights
        w = np.multiply(w, np.exp([float(x) * alpha_m for x in miss2]))
        # Add to prediction
        pred_train = [sum(x) for x in zip(pred_train, 
                                          [x * alpha_m for x in pred_train_i])]
        pred_test = [sum(x) for x in zip(pred_test, 
                                         [x * alpha_m for x in pred_test_i])]
    
    pred_train, pred_test = np.sign(pred_train), np.sign(pred_test)
    # Return error rate in train and test set
    return get_error_rate(pred_train, Y_train), \
           get_error_rate(pred_test, Y_test)

""" PLOT FUNCTION ==========================================================="""
def plot_error_rate(er_train, er_test):
    df_error = pd.DataFrame([er_train, er_test]).T
    df_error.columns = ['Training', 'Test']
    plot1 = df_error.plot(linewidth = 3, figsize = (8,6),
            color = ['lightblue', 'darkblue'], grid = True)
    plot1.set_xlabel('Number of iterations', fontsize = 12)
    plot1.set_xticklabels(range(0,450,50))
    plot1.set_ylabel('Error rate', fontsize = 12)
    plot1.set_title('Error rate vs number of iterations', fontsize = 16)
    plt.axhline(y=er_test[0], linewidth=1, color = 'red', ls = 'dashed')

""" MAIN SCRIPT ============================================================="""
if __name__ == '__main__':
    
#     # Read data
#     x, y = make_hastie_10_2()
#     df = pd.DataFrame(x)
#     df['Y'] = y

#     # Split into training and test set
#     train, test = train_test_split(df, test_size = 0.2)
#     X_train, y_train = train.iloc[:,:-1], train.iloc[:,-1]
#     X_test, y_test = test.iloc[:,:-1], test.iloc[:,-1]
    
    # Fit a simple decision tree first
    clf_tree = DecisionTreeClassifier(max_depth = 3, random_state = 1)
    er_tree = generic_clf(y_train, X_train, y_test, X_test, clf_tree)
    
    # Fit Adaboost classifier using a decision tree as base estimator
    # Test with different number of iterations
    er_train, er_test = [er_tree[0]], [er_tree[1]]
    x_range = range(10, 410, 10)
    for i in x_range:    
        er_i = adaboost_clf(y_train, X_train, y_test, X_test, i, clf_tree)
        er_train.append(er_i[0])
        er_test.append(er_i[1])
    
    # Compare error rate vs number of iterations
    plot_error_rate(er_train, er_test)
````
